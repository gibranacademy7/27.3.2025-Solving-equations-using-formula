```python
import numpy as np

# Define the matrix A and vector B
A = np.array([[1, -1], [3, 2]])
B = np.array([8, 24])

# Compute the determinant
det_A = np.linalg.det(A)

# Compute the inverse if determinant is nonzero
if det_A != 0:
    A_inv = np.linalg.inv(A)
    # Solve for X using the inverse
    X = np.dot(A_inv, B)
else:
    A_inv = None
    X = None

# Display results
print("Determinant of A:", det_A)
if A_inv is not None:
    print("Inverse of A:")
    print(A_inv)
    print("Solution X:")
    print(X)
else:
    print("Matrix A is singular and has no inverse.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the matrix A and vector B
A = np.array([[1, -1], [3, 2]])
B = np.array([8, 24])

# Compute the determinant
det_A = np.linalg.det(A)

# Compute the inverse if determinant is nonzero
if det_A != 0:
    A_inv = np.linalg.inv(A)
    # Solve for X using the inverse
    X = np.dot(A_inv, B)
else:
    A_inv = None
    X = None

# Display results
print("Determinant of A:", det_A)
if A_inv is not None:
    print("Inverse of A:")
    print(A_inv)
    print("Solution X:")
    print(X)
else:
    print("Matrix A is singular and has no inverse.")

# Plot the system of equations
def plot_system():
    x = np.linspace(-5, 10, 400)
    y1 = x - 8  # From equation x - y = 8
    y2 = (24 - 3*x) / 2  # From equation 3x + 2y = 24

    plt.figure(figsize=(8, 6))
    sns.set(style="whitegrid")
    plt.plot(x, y1, label="x - y = 8", color="blue")
    plt.plot(x, y2, label="3x + 2y = 24", color="red")
    
    # Plot intersection point
    if X is not None:
        plt.scatter(X[0], X[1], color="black", zorder=3, label=f"Solution ({X[0]:.2f}, {X[1]:.2f})")
    
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.legend()
    plt.title("Graphical Representation of the System of Equations")
    plt.show()

plot_system()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define matrix A and vector B
A = np.array([[6, 1], [2, 3]])
B = np.array([12, 20])

# Compute determinant
det_A = np.linalg.det(A)

# Compute inverse of A
A_inv = np.linalg.inv(A)

# Solve for x and y
solution = np.linalg.solve(A, B)
x, y = solution

# Print results
print(f"Determinant of A: {det_A}")
print(f"Inverse of A:\n{A_inv}")
print(f"Solution (x, y): {solution}")

# Visualization
sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize=(6, 6))

# Plot the lines
x_vals = np.linspace(-5, 5, 100)
y1_vals = (12 - 6*x_vals)
y2_vals = (20 - 2*x_vals) / 3

ax.plot(x_vals, y1_vals, label="6x + y = 12", color='blue')
ax.plot(x_vals, y2_vals, label="3y + 2x = 20", color='red')

# Mark the solution point
ax.scatter(x, y, color='black', zorder=3, label=f"Solution ({x}, {y})")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.legend()
plt.title("Graphical Solution of the System")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define matrix A
A = np.array([[3, -2, -4], 
              [1, 3, -2], 
              [2, -4, 3]])

# Compute determinant
det_A = np.linalg.det(A)

# Compute cofactors matrix
cofactors = np.zeros_like(A, dtype=float)

for i in range(3):
    for j in range(3):
        # Compute minor
        minor = np.delete(np.delete(A, i, axis=0), j, axis=1)
        # Compute determinant of the minor
        cofactors[i, j] = ((-1) ** (i + j)) * np.linalg.det(minor)

# Compute adjugate (transpose of cofactor matrix)
adj_A = cofactors.T

# Compute inverse
inv_A = adj_A / det_A

# Print results
print("Determinant of A:", round(det_A, 2))
print("Adjugate of A:\n", adj_A)
print("Inverse of A:\n", inv_A)

# --- Visualization Setup ---
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Titles and Data
titles = ["Matrix A", "Adjugate Matrix", "Inverse Matrix"]
matrices = [A, adj_A, inv_A]

# Custom Heatmap Plot Function
def plot_matrix(ax, data, title):
    sns.heatmap(data, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, cbar=True, ax=ax)
    ax.set_title(title, fontsize=12)
    
# Plot each matrix
for ax, matrix, title in zip(axes, matrices, titles):
    plot_matrix(ax, matrix, title)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define matrix A
A = np.array([[3, -2, -4], 
              [1, 3, -2], 
              [2, -4, 3]])

# Compute determinant
det_A = np.linalg.det(A)

# Compute cofactors matrix
cofactors = np.zeros_like(A, dtype=float)
for i in range(3):
    for j in range(3):
        minor = np.delete(np.delete(A, i, axis=0), j, axis=1)
        cofactors[i, j] = ((-1) ** (i + j)) * np.linalg.det(minor)

# Compute adjugate (transpose of cofactor matrix)
adj_A = cofactors.T

# Compute inverse
inv_A = adj_A / det_A

# --- 3D Plotting ---
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# X, Y coordinates for matrices
X, Y = np.meshgrid(range(3), range(3))

# Define Z heights for each matrix layer
Z_A = np.full((3, 3), 3)  # A at height z=3
Z_adj = np.full((3, 3), 2)  # Adjugate at height z=2
Z_inv = np.full((3, 3), 1)  # Inverse at height z=1

# Plot surfaces for each matrix
ax.plot_surface(X, Y, Z_A, facecolors=plt.cm.coolwarm((A - A.min()) / (A.max() - A.min())), rstride=1, cstride=1, alpha=0.8)
ax.plot_surface(X, Y, Z_adj, facecolors=plt.cm.coolwarm((adj_A - adj_A.min()) / (adj_A.max() - adj_A.min())), rstride=1, cstride=1, alpha=0.8)
ax.plot_surface(X, Y, Z_inv, facecolors=plt.cm.coolwarm((inv_A - inv_A.min()) / (inv_A.max() - inv_A.min())), rstride=1, cstride=1, alpha=0.8)

# Annotate values on the surfaces
for i in range(3):
    for j in range(3):
        ax.text(i, j, 3.1, f"{A[j, i]:.2f}", color='black', fontsize=10, ha='center')
        ax.text(i, j, 2.1, f"{adj_A[j, i]:.2f}", color='black', fontsize=10, ha='center')
        ax.text(i, j, 1.1, f"{inv_A[j, i]:.2f}", color='black', fontsize=10, ha='center')

# Labels & Titles
ax.set_xlabel("Columns")
ax.set_ylabel("Rows")
ax.set_zlabel("Matrix Type")
ax.set_title("3D Visualization of A, Adjugate(A), and Inverse(A)")

# Customizing axis ticks
ax.set_zticks([1, 2, 3])
ax.set_zticklabels(["Inverse A", "Adjugate A", "Matrix A"])

plt.show()
```
